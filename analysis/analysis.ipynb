```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib.colors import LogNorm

import seaborn as sns
sns.set_theme(font_scale=1.5)
```

```python
h = np.linspace(0, 999, 1000)
E = np.logspace(np.log10(0.250), np.log10(10000), 100)

dep_data = pd.read_csv("../build/electron_dep_test.csv", names=['E'])
ph_dep_data = pd.read_csv("../build/photon_dep_", names=['E'])

sp_data    = pd.read_csv("../build/electron_ene_test.csv", names=E)
ph_sp_data = pd.read_csv("../build/photon_ene_test.csv", names=E)



fig = plt.figure(figsize=(16,8)); plt.subplot(1,2,1); plt.grid(True, which='both')
plt.plot(dep_data.E.values, h);
plt.xlabel('Energy Deposited [keV/km]');
plt.ylabel('Altitude above 1 bar [km]');

plt.xscale('log')

plt.ylim([0, 500])

plt.subplot(1,2,2); plt.grid(True, which='both')
plt.plot(ph_dep_data.E.values, h);
plt.xlabel('Energy Deposited [keV/km]');
plt.ylabel('Altitude above 1 bar [km]');

plt.xscale('log')

plt.ylim([0, 500])


X,Y = np.meshgrid(E, h)

plt.figure(figsize=(16,8)); plt.subplot(1,2,1); plt.grid(True, which='both')
plt.pcolormesh(X, Y, sp_data, norm=LogNorm());
plt.xscale('log');
plt.xlabel('Energy [keV]');
plt.ylabel('Altitude [km]');
plt.colorbar()

plt.subplot(1,2,2); plt.grid(True, which='both')
plt.pcolormesh(X, Y, ph_sp_data, norm=LogNorm());
plt.xscale('log');
plt.xlabel('Energy [keV]');
plt.ylabel('Altitude [km]');
plt.colorbar();
```
